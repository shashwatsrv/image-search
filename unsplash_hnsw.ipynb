```python
import numpy as np

embeddings = np.load("data/unsplash/embeddings.npy")
ids = np.load("data/unsplash/ids.npy", allow_pickle=True)

print(embeddings.shape, len(ids))
```

```python
import faiss

embeddings = embeddings.astype("float32")

d = embeddings.shape[1]

index = faiss.IndexHNSWFlat(d, 32)
index.hnsw.efConstruction = 200

index.add(embeddings)

print("Indexed:", index.ntotal)
```

```python
import os

os.makedirs("data/unsplash", exist_ok=True)

faiss.write_index(index, "data/unsplash/index_hnsw.faiss")

print("Saved HNSW index")
```
